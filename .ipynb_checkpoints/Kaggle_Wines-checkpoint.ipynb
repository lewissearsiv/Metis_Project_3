```python
import sys
print("Python Version:", sys.version)
import statistics
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%config InlineBackend.figure_format = 'svg'
%matplotlib inline
sns.set()
```

```python
df_raw_wine = pd.read_csv('Wine_Data/kaggle_wine.csv')
df_wine = df_raw_wine.iloc[:,1:]
len(df_wine)
```

```python
df_raw_wine.head()
```

```python
df_us = df_wine[df_wine['country'] == 'US']
df_fr = df_wine[df_wine['country'] == 'France']
df_it = df_wine[df_wine['country'] == 'Italy']
df_wine3 = pd.concat([df_us, df_fr,df_it])
df_wine3['country'].value_counts()
```

```python
plt.hist(df_wine3.points)
plt.title('Histogram of Scores');
```
