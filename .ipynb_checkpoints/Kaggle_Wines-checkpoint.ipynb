```python
binary_features = df_chard.iloc[:,10:]
target = df_chard['country']
X_train, X_test, y_train, y_test = train_test_split(binary_features,target,test_size = 0.20)
binary_features.head()
```

Let's find which k works best for a k nearest neighbors classifier.

```python
ks = range(2,15)
enumerated_ks =[]
for i in ks:
    knn_test = KNeighborsClassifier(n_neighbors = i)
    knn_test.fit(X_train,y_train)
    enumerated_ks.append(knn_test.score(X_test,y_test))

plt.plot(range(2,15),enumerated_ks)
plt.xlabel('k-values')
plt.ylabel('Model Scores')
plt.title('knn scores')
plt.vlines(x=[7,11],ymin=min(enumerated_ks)-0.01,ymax=max(enumerated_ks)+0.01,linestyles='dashed', 
           color = 'orange', label=['7','11']);
```

It looks like 7 and 11 are giving us the best fit. For simplicity, lets set our k-neighbors model to k=7.

```python
knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(X_train,y_train)
```

```python
knn.score(X_train,y_train)
```

```python
knn.score(X_test,y_test)
```

```python
fig, ax = plt.subplots(figsize=(7, 7))
plot_confusion_matrix(knn, X_test, y_test, ax=ax)
plt.title('K-Nearest Neighbors Confusion Matrix');
```

The confusion matrix shows us that KNN is working really well for US (shocker because 73.75% of the desriptions are US wines) and isn't too bad at the French wines either! Let's give Logistic Regression a try.

```python
logistic = LogisticRegression(solver='liblinear')
logistic.fit(X_train, y_train);
```

```python
logistic.score(X_train,y_train)
```

```python
logistic.score(X_test,y_test)
```

```python
fig, ax = plt.subplots(figsize=(7, 7))
plot_confusion_matrix(logistic, X_test, y_test, ax=ax)
plt.title('Logistic Regression Confusion Matrix');
```

Let's try a Naive Bayes' for a change of pace. These other algorithms are overfitting a little bit. We'll be using the Bernouli method.

```python
NB_bern = BernoulliNB()
NB_bern.fit(X_train,y_train)
NB_bern.score(X_train,y_train)
```

```python
NB_bern.score(X_test,y_test)
```

Atleast we know this model really doesn't seem to overfit at all!

```python
fig, ax = plt.subplots(figsize=(7, 7))
plot_confusion_matrix(logistic, X_test, y_test, ax=ax)
plt.title('Naive Bayes Confusion Matrix');
```

This is the random forest model. Sometimes this bad larry works really well.

```python
forest=RandomForestClassifier(n_estimators=100)
forest.fit(X_train,y_train)
y_pred=forest.predict(X_test)
forest.score(X_train,y_train)
```

```python
forest.score(X_test,y_test)
```

This looks really good!

```python
fig, ax = plt.subplots(figsize=(7, 7))
plot_confusion_matrix(forest, X_test, y_test, ax=ax)
plt.title('Random Forest Confusion Matrix');
```

This one looks like the best so far!! Let's go ahead and check the ROC curves and the AUC scores for all of the classifiers. Then we can have a true comparison.

```python
y_test_bin = y_test.map({'France': 0, 'US': 1})
fpr, tpr, thresholds = roc_curve(y_test_bin, knn.predict_proba(X_test)[:,1])
fprlog, tprlog, thresholdslog = roc_curve(y_test_bin, logistic.predict_proba(X_test)[:,1])
fprnb, tprnb, thresholdsnb = roc_curve(y_test_bin, NB_bern.predict_proba(X_test)[:,1])
fprrf, tprrf, thresholdsrf = roc_curve(y_test_bin, forest.predict_proba(X_test)[:,1])
```

```python
plt.plot(fpr, tpr,lw=2)
plt.plot(fprlog, tprlog,lw=2)
plt.plot(fprnb, tprnb,lw=2)
plt.plot(fprrf, tprrf,lw=2)
plt.plot([0,1],[0,1],c='violet',ls='--')
plt.xlim([-0.05,1.05])
plt.ylim([-0.05,1.05])


#plt.xlabel('False positive rate')
#plt.ylabel('True positive rate')
plt.title('ROC Scores');
plt.legend(['KNN','Logit', 'Naive Bayes', 'Random Forest'])
print("KNN ROC AUC score = ", roc_auc_score(y_test, knn.predict_proba(X_test)[:,1]))
print("Logit ROC AUC score = ", roc_auc_score(y_test, logistic.predict_proba(X_test)[:,1]))
print("Naive Bayes ROC AUC score = ", roc_auc_score(y_test, NB_bern.predict_proba(X_test)[:,1]))
print("Random Forest ROC AUC score = ", roc_auc_score(y_test, forest.predict_proba(X_test)[:,1]))
```

It seems like Random Forest not only had the best score on the test data, but its ROC area under the curve is by far the highest. This is going to be the model we will use moving forward.
